```python
from scipy.stats.stats import pearsonr


import numpy as np
import matplotlib.pyplot as plt
data = np.genfromtxt("https://raw.githubusercontent.com/anaderi/lhcb_trigger_ml/master/hep_ml/datasets/dalitzplot/bkgd.csv", skip_header=1, delimiter="\t")

x[j] = [data.T[j] for j in range(12)]
    
rez = np.array([pearsonr(z, y)[0] for z in x[:9] for y in x[:9]])
plt.imshow(rez.reshape(9, 9), cmap='Reds', interpolation = 'None')

ax = plt.gca()
plt.colorbar()
plt.xticks(np.arange(9), ('M2AB', 'M2AC', 'XA', 'XB', 'XC', 'Y1', 'Y2', 'Y3', 'Y4'), rotation = 'vertical')
plt.yticks(np.arange(9), ('M2AB', 'M2AC', 'XA', 'XB', 'XC', 'Y1', 'Y2', 'Y3', 'Y4'), rotation = 'horizontal')
ax.invert_yaxis()
plt.show()
```
