```python
from mpl_toolkits.mplot3d import Axes3D

import numpy as np
import matplotlib.pyplot as plt
data = numpy.genfromtxt("https://raw.githubusercontent.com/anaderi/lhcb_trigger_ml/master/hep_ml/datasets/dalitzplot/bkgd.csv", skip_header=1, delimiter="\t")

ox = data.T[0]
oy = data.T[1]
fig = plt.figure()
ax = Axes3D(fig)
    
hist, x, y = np.histogram2d(ox, oy, bins = 40, range = [[0, 0.8], [0, 0.8]])
x, y = np.meshgrid(x[:40], y[:40])
surf = ax.plot_surface(x[:40], y[:40], hist, rstride=1, cstride=1, cmap='Blues', linewidth = 0)
fig.colorbar(surf, shrink = 0.8, aspect = 6)

ax.view_init(elev = 60, azim = 30)
    
plt.show()
```
